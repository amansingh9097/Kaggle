# Problem Statement

To analyse the AirBnb & Zillow data of listing of houses and their property values and determine which areas (by their zipcode) are worth invest.

# Assumptions

- The occupancy rate of the houses is 75\%.
- The time value of money discount rate is 0% (i.e. \$1 today is worth the same 100 years from now).
- All properties and all square feet within each locale can be assumed to be homogeneous (i.e. a 1000 square foot property in a locale such as Bronx or Manhattan generates twice the revenue and costs twice as much as any other 500 square foot property within that same locale.)
- The rate of appreciation of the property and the rate of appreciation of the house will remain same over a period of time.
- As the company is only interested in short-term rentals, therefore, the monthly/weekly discounted rates are ignored and is assumed to happen on day by day rates.
- Within any area (zipcode) the rental prices and property prices are assumed to be follow a similar distribution.
- Cleaning fees are not added as part of rental prices, as in most of the places the cleaning fee is only charged when a person vacat's or leaves the rental house. Some apartments provided cleaning service and is thus charged at those places as part of monthly rent. Now as this trend is not consistent in real life practice and varries from place to place, therefore, cleaning fee are not included as part of rental prices.
- Only 2 bedrooms flats are taken into consideration for the analysis.
- Assuming the last reported price (after scrapping of the dataset) of a property is the latest/recent available price of the house.

# Data Preparation
## identifying the zipcode attribute in both datasets & slicing up instances of New York

```python
# rename Region name to zipcode  
zillow.rename(columns={'RegionName':'zipcode'}, inplace=True)

# slice up instances corresponding to Newyork only
zillow = zillow.loc[zillow['City']=='New York']
```

```python
# Convert zipcode to int
zillow['zipcode'] = zillow['zipcode'].astype(int)
```

## some regex corrections

```python
# remove '$' and ',' from all prices fields
airbnb['price'] = airbnb['price'].astype(str).str.replace('$','').str.replace(',','')
airbnb['price'] = pd.to_numeric(airbnb['price'])
airbnb['weekly_price'] = airbnb['weekly_price'].astype(str).str.replace('$','').str.replace(',','')
airbnb['weekly_price'] = airbnb['weekly_price'].astype(float)
airbnb['monthly_price'] = airbnb['monthly_price'].astype(str).str.replace('$','').str.replace(',','')
airbnb['monthly_price'] = airbnb['monthly_price'].astype(float)
airbnb['cleaning_fee'] = airbnb['cleaning_fee'].astype(str).str.replace('$','').str.replace(',','')
airbnb['cleaning_fee'] = airbnb['cleaning_fee'].astype(float)
```

## imputing missing values

```python
# imputing the missing prices: weekly, monthly & cleaning fees
airbnb['weekly_price'].fillna(value=airbnb['price']*7, inplace=True)
airbnb['monthly_price'].fillna(value=airbnb['price']*30, inplace=True)
airbnb['cleaning_fee'].fillna(value=airbnb['cleaning_fee'].mean(), inplace=True)
```

## Standardizing the zipcodes
A look at unique zip codes found many with additional 4 digits appended to the usual 5 digits. These digits are mainly for USPS mail sorting. For the purposes of this analysis, will be keeping only the first 5 digits zipcode so as to have a `proper standard among the zipcodes`

```python
# remove 4 digit route numbers from zip codes
airbnb['zipcode'] = airbnb['zipcode'].astype(str).str.replace('\n','-')
airbnb['zipcode'] = airbnb['zipcode'].str.split('-').str[0]
airbnb['zipcode'] = pd.to_numeric(airbnb['zipcode'], errors='coerce', downcast='integer').fillna(0).astype('int')
airbnb = airbnb.loc[airbnb['zipcode'] > 0]
```

## slicing up instances pertaining to only 2 bedrooms instances, apartments & short rental terms

```python
# slicing up to have only 2 bedrooms instances
airbnb = airbnb.loc[airbnb['bedrooms'] == 2.0]
airbnb = airbnb.loc[airbnb['room_type'] == 'Entire home/apt']
airbnb = airbnb.loc[airbnb['maximum_nights'] <= 28] # short rental terms
```

## creating a method of all future data preprocessing

for all future explorations, directly use the below method for all the preprocessing in one-go

```python
def prepare_data(airbnb, zillow):
    """
    data preprocessing method specifically for 
    cleaning and preprocessing the airbnb & zillow datasets
    """
    zillow.rename(columns={'RegionName':'zipcode'}, inplace=True)
    zillow = zillow.loc[zillow['City']=='New York']
    zillow['zipcode'] = zillow['zipcode'].astype(int)
    airbnb['price'] = airbnb['price'].astype(str).str.replace('$','').str.replace(',','')
    airbnb['price'] = pd.to_numeric(airbnb['price'])
    airbnb['weekly_price'] = airbnb['weekly_price'].astype(str).str.replace('$','').str.replace(',','')
    airbnb['weekly_price'] = airbnb['weekly_price'].astype(float)
    airbnb['monthly_price'] = airbnb['monthly_price'].astype(str).str.replace('$','').str.replace(',','')
    airbnb['monthly_price'] = airbnb['monthly_price'].astype(float)
    airbnb['cleaning_fee'] = airbnb['cleaning_fee'].astype(str).str.replace('$','').str.replace(',','')
    airbnb['cleaning_fee'] = airbnb['cleaning_fee'].astype(float)
    airbnb['weekly_price'].fillna(value=airbnb['price']*7, inplace=True)
    airbnb['monthly_price'].fillna(value=airbnb['price']*30, inplace=True)
    airbnb['cleaning_fee'].fillna(value=airbnb['cleaning_fee'].mean(), inplace=True)
    airbnb['zipcode'] = airbnb['zipcode'].astype(str).str.replace('\n','-')
    airbnb['zipcode'] = airbnb['zipcode'].str.split('-').str[0]
    airbnb['zipcode'] = pd.to_numeric(airbnb['zipcode'], errors='coerce', downcast='integer').fillna(0).astype('int')
    airbnb = airbnb.loc[airbnb['zipcode'] > 0]
    airbnb = airbnb.loc[airbnb['bedrooms'] == 2.0]
    airbnb = airbnb.loc[airbnb['room_type'] == 'Entire home/apt']
    airbnb = airbnb.loc[airbnb['maximum_nights'] <= 28]
    return airbnb, zillow


airbnb, zillow = prepare_data(airbnb, zillow)
```

## Estimating the most profitable zipcodes from Investment perspective
in order to determine the property price appreciation-rate(growth-rate), let's look at the growth-rate over past 6 years.<br>Appreciation rate is the rate of increase in the value of the property over time based on the house price value and the rent given.

Here, assuming that the appreciation rate for the past 6 years would remain as the same rate even for the future 6 years.<br>
Eg., the value of a house is 1 million in 2011 and increased to about 2 million in 2017, then the growth-rate would be
$$ growth\ rate = \frac{present\ value - past\ value}{past\ value}\ * 100 $$

therefore, $growth\ rate = \frac{2\ million - 1\ million}{1\ million}\ * 100 = 100\%$

Now applying the same logic to determine the future value of the same house in 2023 when it was 2 million in 2017,

$$ future\ value = present\ value\ * (1 + growth\ rate) $$

ie., $house\ value\ in\ 2023\ = house\ value\ in\ 2017\ (2\ million)\ * (1 + growth\ rate) = 2\ million\ * (1 + 1) = 4\ million$

so, the house appreciation value will be $ (4\ million - 2\ million) = 2\ million $

```python
# latest property price - 'date'
zillow.columns[-1]
```

```python
# calculate the growth rate
zillow_df = zillow_prices[['zipcode','2011-06','2017-06']].rename(columns={'2011-06':'price_2011', '2017-06':'price_2017'})
```

```python
airbnb_df = airbnb.groupby(['zipcode'])['price'].median().reset_index().rename(columns={'price': 'latest_rent'})
```

```python
df = pd.merge(airbnb_df,zillow_df, on='zipcode', how='inner')
df.head()
```

Lets try calculating the appreciation rate over 6 years. `Appreciation rate of the house` is the increase in the value of the house price over time based on the house value and the rent given.

From here on, keeping in mind the pre-stated assumption that at any time the occupancy rate of the houses across the zipcodes is 75%

### the Appreciation Rate

```python
# assumed occupancy rate
OCCUPANCY_RATE = 0.75

# Calculate appreciation rate over 6 year period
df['appreciation_rate'] = (df.price_2017 - df.price_2011)/(df.price_2011)

# Income from Home Value appreciation:
# Income from House Value appreciation = Predicted house value for 2023 - House value in 2017
df['price_2023'] = df.price_2017 * (1 + df.appreciation_rate)
df['price_2023'] = df['price_2023'].astype(int)
df['house_value_appreciation'] = df.price_2023 - df.price_2017
df['house_value_appreciation'] = df['house_value_appreciation'].astype(int)
```

### Estimating rent in next 6 years & profit from rent

```python
# Income from Rent:
# Income from AirBnb rentals = (Daily rent in 2017 + Predicted daily rent in 2023) / 2 * (Occupancy rate) * (365 * 6)
df['rent_2023'] = df.latest_rent * (1 + df.appreciation_rate)
df['rent_2023'] = df['rent_2023'].astype(float).round(2)
df['profit_from_rent'] = df[['latest_rent', 'rent_2023']].mean(axis=1) * OCCUPANCY_RATE * (365 * 6)
df['profit_from_rent'] = df['profit_from_rent'].astype(float).round(2)
```

### total profit & ROI %

```python
# Total profit and ROI %age:
df['total_profit'] = df.profit_from_rent + df.house_value_appreciation
df['total_profit'] = df['total_profit'].astype(int)
df['roi_percent'] = (df.total_profit / df.price_2017) * 100
df['roi_percent'] = df['roi_percent'].astype(float).round(2)
```

### the Breakeven Period

```python
# breakeven period
df['breakeven'] = 1 / (df.total_profit / df.price_2017)
df['breakeven'] = df['breakeven'].astype(float).round(2)

# Sort the data by RPO
df = df.sort_values(by=['roi_percent'], ascending=[False])
df['occupancy'] = 0.75
df.head()
```

### estimating the CAGR values

$$ CAGR\ value = (\frac{EB}{BB})^{\frac{1}{n}} -1 $$

and $n$ is the number of years

```python
df['CAGR'] = (df['price_2023']/df['price_2017'])**(1/6) - 1
df['CAGR'] = df['CAGR'].astype(float).round(3)
```

# BONUS - Estimating the right 'occupancy rate'

The above analysis is based on the pre-stated assumption that the houses have 75% occupancy rate at any time of the year, across the zipcodes. But for a moment, lets try evaluating the right occupancy rate among the zipcodes.

## The hypothesis
`Our Hypothesis`: Let us assume that every person who stays in the house has to write a review of the location and somehow, the total *review_scores_location* is directly proportional to the *occupancy rate* of the house.

The better the score of a location the more occupied houses there are throughout the year, ie. more occupancy rate and vice versa.

As can be seen, out of 1179 instances of zipcodes, we have review_score_location for only 947 of them.

Lets use some impute the review_score_location for the remaining 232 of them with the mean review_scores_location of all the scores from a particular neighbourhood_group.

```python
print("min review score of any location: ", airbnb.review_scores_location.min())
print("max review score of any location: ", airbnb.review_scores_location.max())
```

```python
airbnb.review_scores_location.fillna(value=airbnb.review_scores_location.mean(), inplace=True)
```

```python
airbnb.review_scores_location.isna().sum()
```

## Setting bins based on location's review scores
Following the same assumption of review of location, lets say 
- if a location received review score less than 5.0 has about 55% occupancy rate. 
- Similarly, locations with 5.0 to 6.5 have about 65% occupancy rate; 
- locations with review score of 6.6 to 8.0 have about 70% occupancy rate and 
- any location with 8.1 or more have an occupancy rate of 75%.

Therefore, the new occupancy rates specific to the locations w.r.t. their location's review scores would be,

```python
airbnb.loc[airbnb['review_scores_location'] <= 5, 'occupancy'] = 0.55
airbnb.loc[(airbnb['review_scores_location'] > 5) & (airbnb['review_scores_location'] <= 6.5), 'occupancy'] = 0.65
airbnb.loc[(airbnb['review_scores_location'] > 6.5) & (airbnb['review_scores_location'] < 8), 'occupancy'] = 0.70
airbnb.loc[airbnb['review_scores_location'] >= 8, 'occupancy'] = 0.75
airbnb['review_scores_location'] = airbnb['review_scores_location'].astype(float).round(1)
```

## Re-evaluating the most profitable zipcodes from Investment perspective
Now lets re-evaluate the top 10 most profitable zipcodes based on ROI%, breakeven period and CAGR scores:

```python
#TODO: decimals calculate growth rate
zillow_df = zillow_prices[['zipcode','2011-06','2017-06']].rename(columns={'2011-06':'price_2011', '2017-06':'price_2017'})
airbnb_df = airbnb.groupby(['zipcode'])['price'].median().reset_index().rename(columns={'price': 'latest_rent'})
df = pd.merge(airbnb_df,zillow_df, on='zipcode', how='inner')\
.merge(airbnb.groupby(['zipcode'])['occupancy'].median().reset_index(), on='zipcode')

# Calculate appreciation rate over 6 year period
df['appreciation_rate'] = (df.price_2017 - df.price_2011)/(df.price_2011)

# Income from Home Value appreciation:
# Income from House Value appreciation = Predicted house value for 2023 - House value in 2017
df['price_2023'] = df.price_2017 * (1 + df.appreciation_rate)
df['price_2023'] = df['price_2023'].astype(int)
df['house_value_appreciation'] = df.price_2023 - df.price_2017
df['house_value_appreciation'] = df['house_value_appreciation'].astype(int)

# Income from Rent:
# Income from AirBnb rentals = (Daily rent in 2017 + Predicted daily rent in 2023) / 2 * (Occupancy rate) * (365 * 6)
df['rent_2023'] = df.latest_rent * (1 + df.appreciation_rate)
df['rent_2023'] = df['rent_2023'].astype(float).round(2)
df['profit_from_rent'] = ((df['latest_rent'] + df['rent_2023']) * 0.5) * df['occupancy'] * (365 * 6)
df['profit_from_rent'] = df['profit_from_rent'].astype(float).round(2)

# Total profit and ROI %age:
df['total_profit'] = df.profit_from_rent + df.house_value_appreciation
df['total_profit'] = df['total_profit'].astype(int)
df['roi_percent'] = (df.total_profit / df.price_2017) * 100
df['roi_percent'] = df['roi_percent'].astype(float).round(2)

# breakeven period
df['breakeven'] = 1 / (df.total_profit / df.price_2017)
df['breakeven'] = df['breakeven'].astype(float).round(2)

# Sort the data by RPO
df = df.sort_values(by=['roi_percent'], ascending=[False])

# CAGR scores
df['CAGR'] = (df['price_2023']/df['price_2017'])**(1/6) - 1
df['CAGR'] = df['CAGR'].astype(float).round(3)
```
